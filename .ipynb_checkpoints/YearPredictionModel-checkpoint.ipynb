```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('CompleteDataset.csv')
data.head(10)
```

```python
data = data.drop(['name','id','year','GeoLocation'],axis='columns')
```

```python
abs(data.corr()["Timestamp"])[abs(data.corr()["Timestamp"])>0.5]
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
predictors = data[data.columns[data.columns != 'Timestamp']]
predictors

target = data['Timestamp']
target

predictors = (predictors - predictors.mean()) / predictors.std()
target = (target - target.mean()) / target.std()
```

```python

def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(50, activation='relu', input_shape=(predictors.shape[1],)))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python

model = regression_model()

model.fit(predictors, target , validation_split=0.3 , epochs=50, verbose=2)
```
