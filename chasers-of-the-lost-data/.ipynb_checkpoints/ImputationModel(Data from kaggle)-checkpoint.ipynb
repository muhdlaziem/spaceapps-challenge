### Since Oribiting Body, Equinox has only only one unique value, we drop together with Orbit Determination Date and Close Approach Date

```python
exp = asteroids.drop(['Neo Reference ID','Name','Orbiting Body','Equinox','Orbit Determination Date','Close Approach Date'],axis='columns')
```

### Impute Aphelion Dist 

```python
abs(exp.corr()["Aphelion Dist"])[abs(exp.corr()["Aphelion Dist"])>0.5]
```

```python
train =exp[["Jupiter Tisserand Invariant","Eccentricity","Semi Major Axis","Orbital Period","Mean Motion","Aphelion Dist"]]
train.corr()
```

#### using deep learning techniques(datawig)

```python
import datawig

# df = datawig.utils.generate_df_numeric(num_samples=200, data_column_name='x', label_column_name='y')         
df_train, df_test = datawig.utils.random_split(train)

#Initialize a SimpleImputer model
imputer = datawig.SimpleImputer(
    input_columns=["Jupiter Tisserand Invariant","Eccentricity","Semi Major Axis","Orbital Period","Mean Motion"], # column(s) containing information about the column we want to impute
    output_column="Aphelion Dist", # the column we'd like to impute values for
    output_path = 'imputer_model' # stores model data and metrics
    )

#Fit an imputer model on the train data
imputer.fit(train_df=df_train, num_epochs=100)

#Impute missing values and return original dataframe with predictions
imputed = imputer.predict(df_test)
             
```

```python
imputed
```

```python
import numpy as np
def rmse(Y, Y_pred): #calculate rmse
    rmse = np.sqrt(sum((Y - Y_pred) ** 2) / len(Y))
    return rmse
def r2_score(Y, Y_pred): #calculate r2
    mean_y = np.mean(Y)
    ss_tot = sum((Y - mean_y) ** 2)
    ss_res = sum((Y - Y_pred) ** 2)
    r2 = 1 - (ss_res / ss_tot)
    return r2
```

```python
print("RMSE: " + str(rmse(imputed["Aphelion Dist"],imputed["Aphelion Dist_imputed"])))
print("R2 Score: " + str(r2_score(imputed["Aphelion Dist"],imputed["Aphelion Dist_imputed"])))
```
